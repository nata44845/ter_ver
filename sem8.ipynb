```python

import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

x = np.array([10,8, 13, 9,11,14, 6,4,12, 7,5])
y = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68 ])

print(np.corrcoef(x,y))

plt.scatter(x, y)
plt.title('r=0.816')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
